### 1. Tokenization
#### Tokenization is the first step in processing text for Large Language Models (LLMs). It is the process of splitting the words and sentences into tokens(smaller units) which can be words, subwords, or characters that a model can interpret and understand.

##### Basic Tokenization using Split

```python
String = "AI in February: 28 Days of Generative AI"
```

```python
tokenize = String.split()
tokenize
```

### Using regular expression(re) to perform tokenisation.

```python
import re
```

```python
re_tokenize = re.split(' ', String)
re_tokenize
```

### Using a popular library in NLP(nltk) to perform tokenisation.

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

#### Using Intelligent models like transformers(BERT model)

```python
!pip3 install transformers
```

```python
nltk_tokenize = nltk.word_tokenize(String)
nltk_tokenize
```

```python
from transformers import BertTokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-cased")
Bert_token = tokenizer.tokenize(String)
Bert_token
```

```python
from transformers import BertTokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
Bert_un_token = tokenizer.tokenize(String)
Bert_un_token
```

Showing how Ngrams works and performs tokenisation

```python
from nltk import ngrams
ngrams_1 = ngrams(String.split(), n=1)
ngrams_1
```

```python
for i in ngrams_1:
    print(i)
```

```python
ngrams_2 = ngrams(String.split(), n=2)
ngrams_2
for i in ngrams_2:
    print(i)
```

```python
ngrams_3 = ngrams(String.split(), n=3)
ngrams_3
for i in ngrams_3:
    print(i)
```
